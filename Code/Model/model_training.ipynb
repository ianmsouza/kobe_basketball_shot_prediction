```python
import pandas as pd 
from pycaret.classification import *
import mlflow

# Carregar dados
train_data = pd.read_parquet("../../Data/Processed/base_train.parquet")

# Configurar PyCaret com log do MLflow
exp = setup(data=train_data, target='shot_made_flag', log_experiment=True, experiment_name='Treinamento')

# Treinar regressão logística
lr_model = create_model('lr')
lr_metrics = pull()
print("Métricas do modelo LR:")
print(lr_metrics)
# Obter o F1 médio
lr_f1 = lr_metrics.loc["Mean", "F1"]
mlflow.log_metric("f1_lr", lr_f1)

# Treinar árvore de decisão
dt_model = create_model('dt')
dt_metrics = pull()
print("Métricas do modelo DT:")
print(dt_metrics)
# Obter o F1 médio
dt_f1 = dt_metrics.loc["Mean", "F1"]
mlflow.log_metric("f1_dt", dt_f1)

# Selecionar o melhor modelo (quanto maior o F1, melhor)
if lr_f1 > dt_f1:
    final_model = lr_model
else:
    final_model = dt_model

# Salvar modelo final
save_model(final_model, "artifacts/modelo_final")
```
