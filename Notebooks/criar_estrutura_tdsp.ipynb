**Definindo a estrutura de diretórios conforme o TDSP**
<br><br>


```python
import os

def criar_estrutura_tdsp(raiz_projeto):
    # Estrutura exatamente conforme a imagem enviada (FIG 2 do TDSP)
    estrutura = [
        'Data/Raw',
        'Data/Processed',
        'Data/Modeling',
        'Docs/Project',
        'Docs/DataReport',
        'Docs/Model',
        'Code/DataPrep',
        'Code/Model',
        'Code/Operationalization',
        'Notebooks'
    ]

    # Criação dos diretórios
    for diretorio in estrutura:
        caminho = os.path.join(raiz_projeto, diretorio)
        os.makedirs(caminho, exist_ok=True)
        print(f'Diretório criado: {caminho}')

    # Criação dos arquivos obrigatórios na raiz e Setup
    arquivos = {
        "README.md": "# Projeto Engenharia de ML\n",
        "requirements.txt": "# pacotes necessários\n"
    }

    for arquivo, conteudo in arquivos.items():
        caminho_arquivo = os.path.join(raiz_projeto, arquivo)
        with open(caminho_arquivo, "w") as f:
            f.write(arquivos[arquivo])
        print(f'Arquivo criado: {caminho_arquivo}')

# Ajuste conforme sua instrução
criar_estrutura_tdsp(r'C:\Users\Ian\PythonProjects\infnet-25E1_3')
```
